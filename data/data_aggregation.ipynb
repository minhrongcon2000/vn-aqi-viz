```python
import pandas as pd
import glob
import numpy as np
```

```python
filenames = glob.glob("*.csv")
```

```python
dfs = []
column_names = set()
for filename in filenames:
    if not filename.startswith("aqi"):
        df = pd.read_csv(filename, na_values=[" "]).fillna(0)
        df.columns = [column.strip() for column in df.columns]
        df['province'] = filename[:-4]
        for column_name in df.columns:
            column_names.add(column_name)
        dfs.append(df)
```

```python
for df in dfs:
    for column_name in column_names:
        if column_name not in df.columns:
            df[column_name] = 0
```

```python
all_data = pd.concat(dfs, axis=0)
```

```python
all_data
```

```python
all_data.loc[all_data["aqi"] == 0, "aqi"] = all_data.loc[all_data["aqi"] == 0, ["pm25", "pm10", "o3", "no2", "so2", "co"]].max(axis=1)
```

```python
all_data
```

```python
all_data.to_csv("aqi.csv", index=False)
```
